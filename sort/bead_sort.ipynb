```python
def bead_sort(sequence: list) -> list:
    if any(not int(x) or x < 0 for x in sequence):
        raise TypeError("Sequence must be a list of non-negative integers")

    if not sequence:
        return sequence

    # Create a list of lists to simulate the beads falling
    print(sequence)
    beads = []
    for num in sequence:
        beads.append([1] * num + [0] * (len(sequence) - num))

    print(beads)
    # Transpose and count the beads
    transposed = list(zip(*beads))
    result = [sum(col) for col in transposed]

    return result
```

```python
print(bead_sort([3, 4, 1, 2, 5]))
```
